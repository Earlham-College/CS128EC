## Graphing Examples

Create a plot to show child **A**'s height, if child **A** was 50.8 cm at birth, 75.7 cm at age 1, 86.8 cm at age 2, 95.2 cm at age 3, and 102.3 cm at age 4. Make sure to label the axes.

````python
age = [0,1,2,3,4] # years
height = [50.8, 75.7, 86.8, 95.2, 102.3] # cm
````

````python
len(age), len(height)
````

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(age, label='Age')
plt.plot(height, label='Height')
plt.legend()
plt.show()
````

````python
plt.plot(age, height) 
plt.ylabel('HEIGHT')
plt.xlabel('AGE')
plt.show()
````

Modify the plot above to include the information that child **A** was 67.5 cm at 6 months, 82.2 cm at age 1.5, 91.1 cm at age 2.5, and 98.8 cm at age 3.5.

````python
def appendA():
    ageN = 0
    while ageN == 0: 
        ageN = float(input("New height [months] "))
        age.append(ageN)
    return age
appendA()
# age.append(6)
# age.append(18)
# age.append(30)
# age.append(42)
#age
````

````python
def appendH():
    heightN = 0
    while heightN == 0: 
        heightN = float(input("New height [cm]"))
        height.append(heightN)
    return height
appendH()
````

````python
plt.plot(age, height) 
plt.ylabel('HEIGHT')
plt.xlabel('AGE')
plt.show()
````
